```python
import numpy as np

# ---- Calibration constants for HuGaDB IMUs ----
# HuGaDB stores raw 16-bit integer values. Per their docs:
# - Accelerometer: divide by ~16384 to get units of g
# - Gyroscope: divide by ~131 to get degrees/second
ACC_SCALE = 16384.0
GYRO_SCALE = 131.0

dt = 1.0 / fs  # time step in seconds

def estimate_segment_angle(acc_x, acc_y, acc_z, gyro_x, alpha=0.98):
    """
    Estimate the pitch angle of a body segment in the sagittal plane (the
    plane you'd see from the side while walking).

    Uses a complementary filter:
      - Gyroscope integration tracks fast motion accurately but drifts.
      - Accelerometer gives an absolute tilt reference from gravity but is
        noisy during dynamic motion.
      - We blend them: alpha=0.98 trusts the gyro short-term, the
        accelerometer corrects long-term drift.
    """
    # Scale raw sensor values to physical units
    ax = acc_x / ACC_SCALE
    ay = acc_y / ACC_SCALE
    az = acc_z / ACC_SCALE
    gx_deg = gyro_x / GYRO_SCALE  # deg/s

    n = len(ax)
    angle = np.zeros(n)

    # Initial angle from accelerometer (gravity vector projected onto sagittal plane)
    angle[0] = np.degrees(np.arctan2(ay[0], az[0]))

    for i in range(1, n):
        # Gyro integration step
        gyro_angle = angle[i-1] + gx_deg[i] * dt
        # Accelerometer-derived angle
        acc_angle = np.degrees(np.arctan2(ay[i], az[i]))
        # Blend
        angle[i] = alpha * gyro_angle + (1 - alpha) * acc_angle

    return angle

# Compute thigh and shin angles in the sagittal plane
thigh_angle = estimate_segment_angle(
    df['RT_acc_x'].values, df['RT_acc_y'].values, df['RT_acc_z'].values,
    df['RT_gyro_x'].values
)

shin_angle = estimate_segment_angle(
    df['RS_acc_x'].values, df['RS_acc_y'].values, df['RS_acc_z'].values,
    df['RS_gyro_x'].values
)

# Knee angle = angle between thigh and shin
# (when the knee is straight, thigh and shin are aligned, so the difference is ~0;
#  when bent, the shin rotates relative to the thigh)
knee_angle = thigh_angle - shin_angle

print(f"Thigh angle range: {thigh_angle.min():.1f}° to {thigh_angle.max():.1f}°")
print(f"Shin angle range:  {shin_angle.min():.1f}° to {shin_angle.max():.1f}°")
print(f"Knee angle range:  {knee_angle.min():.1f}° to {knee_angle.max():.1f}°")
```

```python
fig, ax = plt.subplots(figsize=(12, 4))

mask = (time >= 10) & (time <= 15)
ax.plot(time[mask], knee_angle[mask], color='steelblue', linewidth=1.5)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Knee angle (°)')
ax.set_title('Right knee flexion during walking (zoomed: 10-15s)')
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from scipy.signal import butter, filtfilt

# 1. Baseline correction: shift so the 5th percentile sits at 0°
#    (5th percentile is more robust than min — won't get thrown by one outlier)
baseline = np.percentile(knee_angle, 5)
knee_corrected = knee_angle - baseline

# 2. Low-pass filter to remove high-frequency noise (anything above 10 Hz is sensor noise, not gait)
nyquist = fs / 2
cutoff = 10.0  # Hz
b, a = butter(N=4, Wn=cutoff / nyquist, btype='low')
knee_smooth = filtfilt(b, a, knee_corrected)

# 3. Plot to confirm it looks right
fig, ax = plt.subplots(figsize=(12, 4))
mask = (time >= 10) & (time <= 15)
ax.plot(time[mask], knee_corrected[mask], color='lightgray', linewidth=1, label='Raw')
ax.plot(time[mask], knee_smooth[mask], color='steelblue', linewidth=2, label='Smoothed')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Knee flexion angle (°)')
ax.set_title('Right knee flexion (baseline-corrected, smoothed)')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Corrected knee angle range: {knee_smooth.min():.1f}° to {knee_smooth.max():.1f}°")
```

```python
from scipy.signal import find_peaks

# Heel-strike events occur near the local minima of the knee angle
# (knee is most extended just before/at heel contact).
# We use the inverted signal so peaks = heel-strikes.
peaks, _ = find_peaks(
    -knee_smooth,
    distance=int(0.7 * fs),  # at least 0.7s between heel-strikes (cadence cap)
    prominence=10            # require a clear dip
)

print(f"Detected {len(peaks)} heel-strike events across the trial")

# Segment the signal into individual gait cycles (heel-strike to next heel-strike)
cycles = []
n_points = 101  # normalize each cycle to 0-100% gait cycle
for i in range(len(peaks) - 1):
    start, end = peaks[i], peaks[i+1]
    cycle = knee_smooth[start:end]
    # Resample to 101 points using linear interpolation
    x_old = np.linspace(0, 100, len(cycle))
    x_new = np.linspace(0, 100, n_points)
    resampled = np.interp(x_new, x_old, cycle)
    cycles.append(resampled)

cycles = np.array(cycles)
print(f"Extracted {len(cycles)} full gait cycles")

# Compute mean and standard deviation across cycles
mean_cycle = cycles.mean(axis=0)
std_cycle = cycles.std(axis=0)

# THE MONEY FIGURE: average knee angle through one gait cycle
fig, ax = plt.subplots(figsize=(9, 5))
gait_pct = np.linspace(0, 100, n_points)

ax.fill_between(gait_pct, mean_cycle - std_cycle, mean_cycle + std_cycle,
                color='steelblue', alpha=0.25, label='±1 SD')
ax.plot(gait_pct, mean_cycle, color='steelblue', linewidth=2.5, label='Mean')

# Annotate key gait events (approximate, for reference)
ax.axvline(0,   color='gray', linestyle='--', alpha=0.4)
ax.axvline(60,  color='gray', linestyle='--', alpha=0.4)
ax.text(0,  ax.get_ylim()[1]*0.95, ' Heel-strike',  fontsize=9, color='gray')
ax.text(60, ax.get_ylim()[1]*0.95, ' Toe-off (~60%)', fontsize=9, color='gray')

ax.set_xlabel('Gait cycle (%)')
ax.set_ylabel('Knee flexion angle (°)')
ax.set_title(f'Right knee flexion across {len(cycles)} gait cycles\n'
             f'HuGaDB Subject 01, walking trial')
ax.legend(loc='upper right')
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('knee_angle_gait_cycle.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Peak knee flexion: {mean_cycle.max():.1f}° at {gait_pct[mean_cycle.argmax()]:.0f}% gait cycle")
print(f"Minimum knee angle: {mean_cycle.min():.1f}° at {gait_pct[mean_cycle.argmin()]:.0f}% gait cycle")
```
